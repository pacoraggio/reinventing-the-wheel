# IQR, Lower and Upper Whiskers, Outliers

## Interquartile range IQR

The **interquaritle range (IQR)** is a measure of statistical dispersion which means it is an indicator about how the data are spread. It is defined as the difference between the third and first quartile (or the 75<sup>th</sup> and 25<sup>th</sup> percentile) and so, by definition, it contains $50/%$ of the data. 

```python
data = pd.read_excel('./data/superstore_data.xlsx')
```

```python
data.head(3)
```

```python
data['Region'].unique()
```

```python
df_subcategory = (data
 .groupby(['Region', 'Sub-Category'])
 .agg(
     Sales = ('Sales', sum_and_round),
     Profit = ('Profit', sum_and_round),
     Quantity = ('Quantity', sum_and_round),
     Discount = ('Discount', mean_and_round)
 )
 .reset_index()
)
```

```python
df_subcategory.head()
```

```python
obs = 'Profit'

box = (alt
       .Chart(df_subcategory)
       .mark_boxplot(size = 25, color = 'silver')
       .encode(
           x = alt.X('Profit:Q'),
           y = 'Region'
       )
      )

box.properties(
    title = 'Sales by Sub-category box plot',
    width = 620,
    height = 150
)
```
